In the Notebook, we will explore how to create subplots using Matplotlib in Python. Subplots allow us to display multiple plots in a single figure, making it easier to compare different datasets or visualize various aspects of the same dataset side by side.

> Subplots produce a figure of size 640x480 with 2 Axes objects arranged in a 1x2 grid (2D array).

```python
import matplotlib.pyplot as plt
import numpy as np
# Let's create a figure with 1 row and 2 columns of subplots and print the figure size and axes objects
# to understand the structure.
figure, axes = plt.subplots(nrows=1, ncols=2)
print(f"Figure size: {figure.get_size_inches() * figure.dpi} pixels") # Convert size from inches to pixels, dpi is dots per inch
print(f"Axes objects: {axes}, shape: {axes.shape}") # axes is a 2D array with shape (1, 2)
```

```python
# Now, let's create some sample data to plot in our subplots.
x = np.array([1, 2, 3, 4, 5])
y1 = x ** 2  # quadratic growth
y2 = x ** 3 # cubic growth
y3 = np.log10(x)  # logarithmic growth

figure, axes = plt.subplots(2,2) # subplots with 2 rows and 2 columns

# Plotting the data in the subplots
axes[0, 0].plot(x, y1, 'r-')  # First subplot (top-left)
axes[0, 0].set_title('y = x^2')  # Title for first subplot
axes[0, 1].plot(x, y2, 'b-')  # Second subplot (top-right)
axes[0, 1].set_title('y = x^3')  # Title for second subplot
axes[1, 0].plot(x, y3, 'g-')  # Third subplot (bottom-left)
axes[1, 0].set_title('y = log10(x)')  # Title for the third subplot
axes[1, 1].axis('off')  # The fourth subplot (bottom-right) is turned off
figure.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

### Customizing Subplots
We can customize the size of the figure, add titles, and adjust the layout to make the plots more readable.

```python
# Sample data
x = np.linspace(0, 10, 100)
y1 = x ** 2
y2 = np.log10(x + 1)  # Adding 1 to avoid log(0)
y3 = np.exp(x / 3)  # Exponential growth
# Create subplots with a custom figure size
figure, axes = plt.subplots(2, 2, figsize=(10, 8))
# Plotting the data
axes[0, 0].plot(x, y1, 'r-')
axes[0, 0].set_title('y = x^2')
axes[0, 1].plot(x, y2, 'b-')
axes[0, 1].set_title('y = log10(x + 1)')
axes[1, 0].plot(x, y3, 'g-')
axes[1, 0].set_title('y = exp(x / 3)')
axes[1, 1].axis('off')  # Turn off the fourth subplot
figure.suptitle('Various Mathematical Functions', fontsize=16)  # Overall title
figure.tight_layout()
plt.show()
```
