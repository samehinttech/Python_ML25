## Scatter Plots
> Scatter plot is a statistical visualization that displays individual data points on a two-dimensional coordinate system. Each point represents a pair of values—one for the x-axis and one for the y-axis—typically corresponding to two continuous variables.

> Scatter plots are used to reveal the relationship between variables and are especially useful for detecting patterns, correlations, clusters, and outliers within the data. They are a fundamental tool in exploratory data analysis (EDA) and quantitative research.
## When to Use
- **Relationship analysis**: Explore correlations between two continuous variables.
- **Distribution visualization**: Understand how data points are spread across the axes.
- **Outlier detection**: Identify unusual data points that deviate from the overall pattern.
- **Clustering**: Observe natural groupings of data points.
- **Trend identification**: Detect trends or patterns in the data.
- **Large datasets**: Visualize large amounts of data without overlapping points.
- **Multivariate analysis**: Use color, size, or shape to represent additional variables.
- **Exploratory Data Analysis (EDA)**: Initial data exploration to understand relationships.
- **Regression analysis**: Visualize the fit of regression models.
- **Scientific research**: Analyze experimental data and relationships.
- **Business analytics**: Examine relationships between business metrics.
- **Quality control**: Monitor processes and identify deviations.
- **Geospatial analysis**: Plot geographic data points.
- **Time series analysis**: Visualize changes over time.
- **Machine learning**: Analyze feature relationships and model performance.
and the list goes on...

````python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility means the random numbers generated will be the same each time we run the code
np.random.seed(38)

# Simulate study hours (0 to 10)
study_hours = np.arange(0, 11)

'''
- Adds random noise to simulate real-life variability.
- np.random.randint(-5, 6) generates random integers between -5 and 5 (inclusive).
- size=study_hours.shape ensures we generate one random number for each study hour.
- Smaller noise range creates a stronger correlation pattern.
- Example: For a study hour of 4 (expected score = 70), the actual score could be anywhere from 65 to 75
'''
exam_scores = 50 + 5 * study_hours + np.random.randint(-5, 6, size=study_hours.shape)

# Plot the scatter plot
plt.scatter(study_hours, exam_scores, color='blue', marker='o')
plt.title("Study Hours vs Exam Scores (Strong Correlation)")
plt.xlabel("Study Hours per Week")
plt.ylabel("Exam Scores")
plt.grid(True, linestyle=':', alpha=0.5)
plt.show()
````

* We can add a linear regression line to the scatter plot to visualize the trend more clearly and find the line of best fit using NumPy's `polyfit` function.
* We need to calculate the **slope (m)** and **y-intercept (b)** of the linear function **y = mx + b**.
* **`np.polyfit(x, y, degree)`** takes:
  * **x values** = `study_hours` (independent variable)
  * **y values** = `exam_scores` (dependent variable)
  * **degree = 1** (linear/first-degree polynomial)
* **`polyfit` returns two values:**
  * **m** = slope (how steep the line is)
  * **b** = y-intercept (where the line crosses the y-axis)
* After calculating **m** and **b**, we create the regression line using: **`regression_line = m * study_hours + b`**
* Then we plot this line on top of the scatter plot to see how well it fits the data.

````python
np.random.seed(38)
exam_scores = 50 + 5 * study_hours + np.random.randint(-5, 6, size=study_hours.shape)
plt.scatter(study_hours, exam_scores, color='blue', marker='o')
plt.title("Study Hours vs Exam Scores (Realistic Spread)")
plt.xlabel("Study Hours per Week")
plt.ylabel("Exam Scores")
plt.grid(True, linestyle=':', alpha=0.5)

# Computes coefficients of a linear fit (degree 1)
m, b = np.polyfit(study_hours, exam_scores, 1)
# Create a regression line
regression_line = m * study_hours + b
# Plot regression line
plt.plot(study_hours, regression_line, color='red', linestyle='--', label='Regression Line')
plt.legend()
plt.show()
````

From what I gathered, setting a seed in NumPy initializes the random number generator to a specific state. This means that every time you set the same seed and generate random numbers, you'll get the same sequence of numbers. This is useful for reproducibility, especially in scenarios like simulations or testing, where you want consistent results. The seed value itself is just an integer that determines the starting point for the sequence of random numbers.
> I tried different seed values and observed how they affected the generated random numbers. Each unique seed produced a different sequence, but using the same seed consistently yielded the same results across multiple runs. This reinforced my understanding of how seeds control the randomness in NumPy's random number generation.
