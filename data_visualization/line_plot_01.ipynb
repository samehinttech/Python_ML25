## Pre-requisites
Familiarity with Python programming and basic knowledge of I/O, data types, and libraries such as NumPy and Pandas.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create sample data for x and y coordinates
x = [2000, 2001, 2002, 2003, 2004, 2005]
y = [3, 8, 1, 10, 5, 12]

# Plot the data
plt.plot(x, y)

# Add title and labels
plt.title("Line Plot")
plt.xlabel("Year")
plt.ylabel("Revenue (in millions)")

# Show the plot
plt.show()
```

### We can use Numpy with the same example, which makes modifying the data before plotting much easier
*Matplotlib works very well with NumPy arrays.*

**Example:** we want to do the following transformations before plotting:
* y_scaled = y * 2 multiplies all elements of the array at once â€” no loop needed.

Transformation is then vectorized, which is much faster and cleaner than looping through Python lists. This is one of the biggest advantages of using np.array with plotting: easy, fast, element-wise operations.

```python
# Create sample data using NumPy arrays
x = np.array([2000, 2001, 2002, 2003, 2004, 2005])
y = np.array([2,8,4,9,6,12])

# scale all y-values by 2 (e.g., revenue growth)
y_scaled = y * 2

# Plot original data, with customized style to underscore difference. alpha stands for transparency level(0 = fully transparent, 0.5 = 50% transparency, 1 = fully opaque, and default is 1)
plt.plot(
    x, y,
    label='Original',
    marker='o',
    linestyle='--',
    color='red',
    alpha=0.5,
    linewidth=1,
    markersize=10,
    markerfacecolor='blue',
    markeredgewidth=1,
    markeredgecolor='black')

# Plot modified data
plt.plot(x, y_scaled,
         label='Scaled by 2',
         marker='x')

# Add title, labels, and legend
plt.title("Line Plot with NumPy Arrays")
plt.xlabel("Year")
plt.ylabel("Revenue (in millions)")
plt.legend()

# Show the plot
plt.show()
```
