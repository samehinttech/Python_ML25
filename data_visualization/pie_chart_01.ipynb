## Pie Charts
A pie chart, also known as a circle chart, represents data in a circular format, where each slice of the pie corresponds to a category's contribution to the whole. The entire pie represents 100% of the data, and each slice's size is proportional to its value relative to the total.
## When to Use
- **Part-to-whole relationships**: Show how individual categories contribute to the total.
- **Limited categories**: Best for datasets with a small number of categories (typically less than 6).
- **Proportional comparisons**: Visualize relative sizes of categories.
## Common Applications
- Market share distribution
- Budget allocation
- Survey results
- Demographic breakdowns
## Matplotlib Functions
- **`plt.pie()`** - Create pie charts with customizable slices.
- **Customization options**: Colors, labels, explode slices, shadows, start angle.
- **`autopct` parameter**: Display percentage values on slices.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# I will use the same data as in the bar chart example
categories = ["Grains", "Vegetables", "Fruits", "Dairy", "Meat"]
values = np.array([25, 40, 30, 20, 35])

# Create a pie chart
plt.pie(values,
        labels=categories,
        autopct='%1.1f%%',  # Display percentage on slices. 1.1 means one digit before and one after the decimal point, %% is to show the % symbol
        #shadow=True,        # Add a shadow for 3D effect
        startangle=90,     # Rotate the start of the pie chart
        colors=['gold', 'lightgreen', 'lightcoral', 'lightskyblue', 'lightpink'],
        explode=(0, 0, 0, 0.1, 0))  # "explode" the 1st slice (Grains)
plt.title("Food Category Consumption Distribution")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
