# Bar Charts

**Bar charts** visualize and compare discrete categories by displaying their magnitude or frequency.

## When to Use
- **Nominal/Ordinal data**: Categories (product types, education levels)
- **Frequency distributions**: Count occurrences per category
- **Comparisons**: Mean, median, totals across groups
- **Categorical vs. numerical relationships**: Category (x-axis) vs. value (height)

## Common Applications
- Which category is highest/lowest?
- How do groups compare?
- What's the frequency distribution?
- Are there notable variations or outliers?

## Matplotlib Types
- **`plt.bar()`** - Vertical bars (standard)
- **`plt.barh()`** - Horizontal bars (for long labels)
- **Grouped** - Compare subcategories
- **Stacked** - Show part-to-whole relationships

Bar charts provide quick, intuitive insights for exploratory data analysis (EDA) and statistical reporting.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create some categorial data
categories = np.array(["Grains", "Vegetables", "Fruits", "Dairy", "Meat"])
values = np.array([25, 40, 30, 20, 35])

# Create a vertical bar chart
plt.bar(categories, values, color='skyblue', edgecolor='black')
#plt.barh(categories, values, color='skyblue', edgecolor='black') # uncomment this line to show a horizontal bar chart
plt.title("Food Category Consumption")
plt.xlabel("Food Categories")
plt.ylabel("Consumption (in kg)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
